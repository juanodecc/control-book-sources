Una soluciÃģn prÃĄctica a este problema es evaluar el parÃĄmetro variable para distintos valores fijos y de esa forma calcular la ubicaciÃģn en el plano s, de los polos y ceros, para deducir la respuesta temporal del sistema.

Y si lo anterior se pudiera hacer para todos los valores de $\ell \in \mathbb{R}$!. Bueno aquÃ­, es donde entra Walter R. Evans y el "MÃĐtodo de Evans", que permite resolver en forma **grÃĄfica** las funciones del tipo $1+\ell~F(s)=0$ donde $F(s)$ es una funciÃģn de transferencia de la forma $F(s)= \frac{N(s)}{D(s)}$ y $\ell \in \mathbb{R}$, notar que de lo anterior es posible expresar lo de la forma $D(s)+\ell~N(s)=0$. A estos grÃĄficos o diagramas, en control, les llamamos Lugar de las raÃ­ces o en ingles Root Locus.

Este resultado lo usaremos para responder bÃĄsicamente dos preguntas: ÂŋCÃģmo puedo elegir el parÃĄmetro $\ell$ para que la respuesta dinÃĄmica del sistema cumpla con los requerimientos de performance necesarios para el sistema a controlar? y ÂŋQuÃĐ efecto tiene sobre la dinÃĄmica de un sistema, la variaciÃģn de un parÃĄmetro $\ell$ de su funciÃģn de transferencia?

Para comenzaremos a responder estas preguntas, analizaremos el siguiente esquema "genÃĐrico" de control a lazo cerrado del sistema de la figura siguiente, donde $k D(s)$ es el controlador del sistema con ganancia $k$, como parÃĄmetro variable, en principio comenzaremos con $k>0$, pero si bien, en la mayorÃ­a de los problemas que veremos $k$ es positivo,no hay restricciones con respecto a al valor de $k$.

La funciÃģn de transferencia a lazo cerrado es:

$$T(s)=\frac{Y(s)}{R(s)}=\frac{k D(s) G(s)}{1+k D(s) G(s)}$$

donde,

$$ G(s) = \frac{N_G(s)}{D_G(s)}=\frac{\prod_{i=1}^{m_G}(s+z_i)}{\prod_{j=1}^{n_G}(s+p_j)}$$ 

y 

$$ D(s) = \frac{N_D(s)}{D_D(s)}=\frac{\prod_{i=1}^{m_D}(s+z_i)}{\prod_{j=1}^{n_D}(s+p_j)}$$

Los polos y ceros de $G(s)$ son la soluciÃģn de $N_G(s)=0$ y $D_G(s)=0$ respectivamente, y los polos y ceros de $D(s)$ son la soluciÃģn de $N_D(s)=0$ y $D_D(s)=0$.

La funciÃģn de transferencia a lazo cerrado se puede representar de la siguiente forma:

$$T(s)=\frac{k N_D(s) N_G(s)}{D_D(s) D_G(s)+k N_D(s) N_G(s)}$$

```{admonition} Resultado importante 
:class: important

En $T(s)$ se puede ver que los **ceros a lazo cerrado no se "mueven"**, es decir son siempre los mismos al cambiar el valor del parÃĄmetro $k$, pero los **polos** a lazo cerrado sÃ­
```

$$
\begin{matrix}
\text{ceros} \quad T(s) &\Longrightarrow & k N_D(s) N_G(s) = 0 \\
\text{polos} \quad T(s) &\Longrightarrow & D_D(s) D_G(s)+k N_D(s) N_G(s)=0
\end{matrix}
$$

en general, podemos decir que si que si tenemos la funciÃģn de transferencia $F(s)$ donde $s$ es la variable compleja y con polos y ceros, tal que:

$$F(s) = \frac{\prod_{i=1}^{m}(s+z_i)}{\prod_{j=1}^{n}(s+p_j)}$$

podemos halla el mÃģdulo de $F(s)$ haciendo,

$$|F(s)|=\frac{\prod \text{mod. ceros}}{\prod \text{mod. polos}}= \frac{\prod_{i=1}^{m}|s+z_i|}{\prod_{j=1}^{n}|s+p_j|}$$

y el argumento de $F(s)$ serÃĄ,

$$ \angle{F(s)}= \sum{ang. ceros}-\sum{ang. polos}= \sum_{i=1}^{m}{\angle{(s+z_i)}}-\sum_{j=1}^{n}{\angle{(s+p_j)}}$$

## Propiedades de el Lugar de las RaÃ­ces

El lugar geomÃĐtrico de las raÃ­ces, es el lugar geomÃĐtrico de valores de $s$ para el cual la ecuaciÃģn caracterÃ­stica $1 + kD(s)G(s) = 0$, ya que el parÃĄmetro $k \in \mathbb{R}$ varÃ­a desde $0 \longrightarrow \infty$.

$$T(s)=\frac{k D(s) G(s)}{1+k D(s) G(s)}$$

Los polos de $T(s)$ existen cuando el polinomio caracterÃ­stico $1 + k D(s)G(s)$ se anula, es decir,

$$ k D(s) G(s) = -1 = 1 \angle{2 \ell +1}\pi$$

con $\ell = 0 \pm 1, \pm 2, \pm 3, \dots$

o lo que es lo mismo:

$$
\left\{\begin{array}{l}
|k D(s) G(s)|=1\\
\angle{k D(s) G(s)} = \angle{2 \ell +1}\pi
\end{array}\right.
$$

Teniendo en cuanta que la funciÃģn compleja la podemos discriminar en su magnitud y fase, y para $k>=0$, podemos arribar a la siguientes definiciones:

```{admonition} CondiciÃģn de fase
:class: important

El lugar geomÃĐtrico de las raÃ­ces de $D(s)G(s)$ es el lugar geomÃĐtrico de puntos en el plano-$s$ donde la fase de $D(s)G(s)$ es 180.

Esto se lo conoce como la condiciÃģn de fase, que significa matemÃĄticamente:

$$\angle G(s) = 180^o+\ell.360^o,\quad \text{con } \ell \text{ entero.}$$

```

### CondiciÃģn de magnitud o de mÃģdulo

El criterio de magnitud es el siguiente que lo obtenemos aplicando el mÃģdulo a la ecuaciÃģn $1+KG(s)=0$. Entonces resulta:

$$\|D(s) G(s)\|=\frac{1}{|k|}$$

y para $k>0$:

$$k=\frac{1}{\|D(s)\| \|G(s)\|}$$

Este criterio nos permite determnar el $k$, una vez escogido una ubicaciÃģn determinada del lugar geomÃĐtrico de las raÃ­ces.

#### Ejemplo de cÃĄlculo de k con la condiciÃģn de mÃģdulo

si $p_1 = p_2 = -5$ un punto del plano s que cumple con la condiciÃģn de fase para la funciÃģn de transferencia $G(s)=\frac{1}{s(s+10)}$

Calculamos el modulo $|G(p_1)|= \|\frac{1}{-5(-5+10)} \| =\frac{1}{25}$ por lo que por la condiciÃģn de mÃģdulo es:

$$k=\frac{1}{|G(p_1)|}= 25$$

Este valor es el que corresponde al ejemplo anterior (ver tabla de valores y raÃ­ces)

#### **Ejemplo:** cÃĄlculo de fase para un punto especifico del plano-s, $s_0$

Tenemos la siguiente funciÃģn de transferencia G(s) a lazo abierto:

$$G(s)=\frac{s+1}{s\left\{\left[\left(s+2\right)^2+4\right]\left(s+4\right)\right\}}$$

:::{figure-md}
<img style="display:block; margin-left: auto; margin-right: auto;" src="fig10.png" width="500" alt="fig10.png">

$G(s)$ en un punto del plano-s genÃĐrico $s_0$

En la figura anterior mostramos con cruces la ubicaciÃģn de los polos de esta funciÃģn de transferencia, y con cÃ­rculos los ceros. Suponemos un punto de prueba ubicado en $ s_o = -1 + 2j $, y realizamos la suma de las contribuciones de las fases para determinar si es o no punto del lugar de raÃ­ces:

$$
\begin{matrix}

& = & 90^o -116.6^o -0^o -76^o - 33.7^o\\
& = & -136.3^o
\end{matrix}
$$

La fase de $ G $ para este punto de prueba vale $ -136.3^o $, y no $ 180^o $ por lo tanto este punto de prueba no pertenece al lugar geomÃĐtrico de las raÃ­ces.
